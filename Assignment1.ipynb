## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Range of x values

```python
x_values = np.arange(-10, 10.1, 0.1)
```

## Function 1: y = wx + b

```python
def func_one(x, w, b):
  return w * x + b
```

## Function 2: y = x^2

```python
def func_two(x):
  return x**2
```

## Function 3: y = 1 / (1 + e^(-x))

```python
def func_three(x):
  return 1 / (1 + np.exp(-x))
```

## Function 4: y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
def func_four(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

## Function 5: g(f(x)) where u = f(x) = wx + b and y = g(u) = 1 / (1 + e^(-x))

```python
def composite_func_five(x, w, b):
  u = func_one(x, w, b)
  return func_three(u)
```

### Function 6: g(f(x)) where f(x) = wx+b, u = f(x) and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
def composite_func_six(x, w, b):
  u = func_one(x, w, b)
  return func_four(u)
```

# Function 7:
- y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x) and g1(u1) = 1 / (1 + e^(-u1))
- y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x) and g2(u2) = 1 / (1 + e^(-u2))
# y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v) and g3(w) = 1 / (1 + e^(-w))

```python
def composite_func_seven(x, w1, b1, w2, b2, w3, w4, b):
  u1 = func_one(x, w1, b1)
  u2 = func_one(x, w2, b2)
  y1 = func_three(u1)
  y2 = func_three(u2)

  v = w3 * y1 + w4 * y2 + b
  return func_three(v)
```

## Defining Constants

```python
w_linear = 2
b_linear = 3

w_sigmoid_on_linear = 0.5
b_sigmoid_on_linear = -1

w_tanh_on_linear = -1
b_tanh_on_linear = 1

w1 = 0.5
b1 = -1
w2 = -0.5
b2 = 1
w3 = 2
w4 = -1
b = 0.5
```

## Calculating values for each function

```python
y_linear = func_one(x_values, w_linear, b_linear)
y_quadratic = func_two(x_values)
y_sigmoid = func_three(x_values)
y_tanh = func_four(x_values)
y_composite_func_five = composite_func_five(x_values, w_sigmoid_on_linear, b_sigmoid_on_linear)
y_composite_func_six = composite_func_six(x_values, w_tanh_on_linear, b_tanh_on_linear)
y_composite_func_seven = composite_func_seven(x_values, w1, b1, w2, b2, w3, w4, b)
```

## Plotting the output

```python
plt.figure(figsize=(12,15))

# Function One
plt.subplot(4,2,1)
plt.plot(x_values, y_linear)
plt.title("Linear Function")
plt.legend()

# Function Two
plt.subplot(4,2,2)
plt.plot(x_values, y_quadratic)
plt.title('Quadratic Function')
plt.legend()

# Function Three
plt.subplot(4,2,3)
plt.plot(x_values, y_sigmoid)
plt.title('Sigmoid Function')
plt.legend()

# Function Four
plt.subplot(4,2,4)
plt.plot(x_values, y_tanh)
plt.title('Tanh Function')
plt.legend()

# Composite Function Five
plt.subplot(4,2,5)
plt.plot(x_values, y_composite_func_five)
plt.title('Composite-5')
plt.legend()

# Composite Function Six
plt.subplot(4,2,6)
plt.plot(x_values, y_composite_func_six)
plt.title('Composite-6')
plt.legend()

# Composite Functon Seven
plt.subplot(4,2,7)
plt.plot(x_values, y_composite_func_seven)
plt.title('Composite-7')
plt.legend()

plt.show()
```
